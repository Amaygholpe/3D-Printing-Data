```python
data_2=data.copy()
obj=LabelEncoder()
data_2["infill_pattern"] = obj.fit_transform(data["infill_pattern"])
data_2["material"] = obj.fit_transform(data["material"])
```

```python
correlation = data_2.corr()
correlation.style.background_gradient(cmap='hot').set_precision(3)
```

seperate the target goal objects from the list

```python
target_data= data_2["material"].values
feature_data= data_2.drop(["material","print_speed","infill_pattern","bed_temperature","layer_height"],axis=1)
```

```python
target_data
```

```python
feature_data
```

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
X = sc.fit_transform(feature_data)
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, target_data, test_size = 0.2, random_state = 0)
```

```python
from sklearn.tree import DecisionTreeClassifier
Classifier = DecisionTreeClassifier(criterion = 'entropy', random_state = 0)
Classifier.fit(X_train, y_train)
```

```python
print('DecisionTreeClassifierModel Train Score is : ' , Classifier.score(X_train, y_train))
print('DecisionTreeClassifierModel Test Score is : ' , Classifier.score(X_test, y_test))
```

```python
y_pred = Classifier.predict(X_test)
print(np.concatenate((y_pred.reshape(len(y_pred),1), y_test.reshape(len(y_test),1)),1))
```

```python
from sklearn.metrics import confusion_matrix, accuracy_score
cm = confusion_matrix(y_test, y_pred)
print(cm)
accuracy_score(y_test, y_pred)
```
